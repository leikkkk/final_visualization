```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
corr = df.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(200, 20, n=200),
    square=True,
    annot=True,
    fmt=".2f",
    annot_kws={'size':8,'weight':'normal', 'color':'#253D24'}
)
ax.set_title('correlation heatmap') 
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
Pregnancies0 = round(df['Pregnancies'][df['Outcome']==0].mean(),1)
Pregnancies1 = round(df['Pregnancies'][df['Outcome']==1].mean(),1)

Glucose0 = round(df['Glucose'][df['Outcome']==0].mean(),1)
Glucose1 = round(df['Glucose'][df['Outcome']==1].mean(),1)

BloodPressure0 = round(df['BloodPressure'][df['Outcome']==0].mean(),1)
BloodPressure1 = round(df['BloodPressure'][df['Outcome']==1].mean(),1)

SkinThickness0 = round(df['SkinThickness'][df['Outcome']==0].mean(),1)
SkinThickness1 = round(df['SkinThickness'][df['Outcome']==1].mean(),1)

Insulin0 = round(df['Insulin'][df['Outcome']==0].mean(),1)
Insulin1 = round(df['Insulin'][df['Outcome']==1].mean(),1)

BMI0 = round(df['BMI'][df['Outcome']==0].mean(),1)
BMI1 = round(df['BMI'][df['Outcome']==1].mean(),1)

DiabetesPedigreeFunction0 = round(df['DiabetesPedigreeFunction'][df['Outcome']==0].mean(),1)
DiabetesPedigreeFunction1 = round(df['DiabetesPedigreeFunction'][df['Outcome']==1].mean(),1)

Age0 = round(df['Age'][df['Outcome']==0].mean(),1)
Age1 = round(df['Age'][df['Outcome']==1].mean(),1)
```

```python
labels = ['no diabetes', 'diabetes']
Pregnancies = [Pregnancies0, Pregnancies1]
Glucose = [Glucose0, Glucose1]
BloodPressure = [BloodPressure0, BloodPressure1]
SkinThickness = [SkinThickness0, SkinThickness1]
Insulin = [Insulin0, Insulin1]   
BMI = [BMI0, BMI1]
DiabetesPedigreeFunction = [DiabetesPedigreeFunction0, DiabetesPedigreeFunction1]
Age = [Age0, Age1]
x = np.arange(2)
total_width, n = 0.8, 8
width = total_width / n
x = x - (total_width - width) / 2

fig, ax = plt.subplots()
rects1 = ax.bar(x- width*4, Pregnancies, width, label='Pregnancies')
rects2 = ax.bar(x - width*3, Glucose, width, label='Glucose')
rects3 = ax.bar(x - width*2, BloodPressure, width, label='BloodPressure')
rects4 = ax.bar(x - width, SkinThickness, width, label='SkinThickness')
rects5 = ax.bar(x , Insulin, width, label='Insulin')
rects6 = ax.bar(x + width , BMI, width, label='BMI')
rects7 = ax.bar(x + width*2, DiabetesPedigreeFunction, width, label='DiabetesPedigreeFunction')
rects8 = ax.bar(x + width*3, Age, width, label='Age')

ax.set_ylabel('value', fontsize=12)

ax.set(ylim =(0, 160))
ax.set_title('Factors associated with diabetes')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(bbox_to_anchor=(1.02,1.0), ncol=1) 

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)
autolabel(rects8)

#fig.tight_layout()

plt.show()
```
